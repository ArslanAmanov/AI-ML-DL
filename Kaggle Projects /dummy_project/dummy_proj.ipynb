```python
import requests

def get_exchange_rate(base_currency, target_currency):
    url = f"https://api.exchangerate-api.com/v4/latest/{base_currency}"
    response = requests.get(url)
    data = response.json()
    return data['rates'][target_currency] 
```

```python
rate = get_exchange_rate('USD', 'THB')
print(f"The exchange rate from USD to THB is {rate}") 
```

```python
import timeit 
from forex_python.converter import CurrencyRates
from datetime import datetime, timedelta

def get_historical_exchange_rates(base_currency, target_currency, start_date, end_date):
    c = CurrencyRates()
    exchange_rates = {}

    current_date = start_date
    while current_date <= end_date:
        try:
            rate = c.get_rate(base_currency, target_currency, current_date)
            exchange_rates[current_date] = rate
        except:
            print(f"Failed to fetch exchange rate for {current_date}")

        current_date += timedelta(days=1)

    return exchange_rates

# Example usage:
base_currency = 'THB'
target_currency = 'USD'
start_date = datetime(2019, 1, 1)
end_date = datetime(2022, 12, 30)

exchange_rates = get_historical_exchange_rates(base_currency, target_currency, start_date, end_date)

for date, rate in exchange_rates.items():
    print(f"Exchange rate from {base_currency} to {target_currency} on {date}: {rate}")


start_time = timeit.default_timer()
end_time = timeit.default_timer()
elapsed_time = end_time - start_time
print(f"Elapsed time: {elapsed_time} seconds")
```
